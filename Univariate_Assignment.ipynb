#  Measure of location of Data

```python

descriptive = pd.DataFrame(index=["mean","median","Mode","Q1:25%","Q2:50%", "Q3:75%","Q4:100%","IQR","1.5 Rule","Greater","Lesser","kurtosis","skew","var","min","max"],columns=quan)
```

```python
descriptive
```

```python
for cname in quan:
    descriptive.loc["mean", cname] = dataset[cname].mean()
    descriptive.loc["median", cname] = dataset[cname].median()
    descriptive.loc["Mode", cname] = dataset[cname].mode()[0]

    descriptive.loc["Q1:25%", cname] = dataset[cname].quantile(0.25)
    descriptive.loc["Q2:50%", cname] = dataset[cname].quantile(0.50)
    descriptive.loc["Q3:75%", cname] = dataset[cname].quantile(0.75)
    descriptive.loc["Q4:100%", cname] = dataset[cname].max()

    descriptive.loc["IQR", cname] = (
        descriptive.loc["Q3:75%", cname] -
        descriptive.loc["Q1:25%", cname]
    )

    descriptive.loc["1.5 Rule", cname] = (
        1.5 * descriptive.loc["IQR", cname]
    )

    descriptive.loc["Greater", cname] = (
        descriptive.loc["Q3:75%", cname] +
        descriptive.loc["1.5 Rule", cname]
    )

    descriptive.loc["Lesser", cname] = (
        descriptive.loc["Q1:25%", cname] -
        descriptive.loc["1.5 Rule", cname]
    )

    descriptive.loc["kurtosis", cname] = dataset[cname].kurt()
    descriptive.loc["skew", cname] = dataset[cname].skew()
    descriptive.loc["var", cname] = dataset[cname].var()
    descriptive.loc["min", cname] = dataset[cname].min()
    descriptive.loc["max", cname] = dataset[cname].max()
    
    
```

```python
descriptive
```

```python
descriptive
```

```python
lesser =[]
greater=[]
    
```

```python
for cname in quan:
    if descriptive.loc["min", cname] < descriptive.loc["Lesser", cname]:
        lesser.append(cname)

    if descriptive.loc["max", cname] > descriptive.loc["Greater", cname]:
        greater.append(cname)
```

```python
for cname in lesser:
    dataset.loc[
        dataset[cname] < descriptive.loc["Lesser", cname],
        cname
    ] = descriptive.loc["Lesser", cname]

for cname in greater:
    dataset.loc[
        dataset[cname] > descriptive.loc["Greater", cname],
        cname
    ] = descriptive.loc["Greater", cname]
```
