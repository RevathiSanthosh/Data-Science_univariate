#  Measure of location of Data

```python

descriptive = pd.DataFrame(index=["Mean","Median","Mode","Q1:25%","Q2:50%", "Q3:75%","Q4:100%","IQR","1.5 Rule","Greater","Lesser"],columns=quan)
```

```python
descriptive
```

```python
for cname in quan :
    descriptive[cname]["Mean"] = dataset[cname].mean()
    descriptive[cname]["Median"] = dataset[cname].median()
    descriptive[cname]["Mode"] = dataset[cname].mode()[0]
    descriptive[cname]["Q1:25%"] = dataset.describe()[cname]["25%"]
    descriptive[cname]["Q2:50%"] = dataset.describe()[cname]["50%"]
    descriptive[cname]["Q3:75%"] = dataset.describe()[cname]["75%"]
    descriptive[cname]["Q4:100%"] = dataset.describe()[cname]["max"]
    descriptive[cname]["IQR"] = descriptive[cname]["Q3:75%"] - descriptive[cname]["Q1:25%"]
    descriptive[cname]["1.5 Rule"] = 1.5*descriptive[cname]["IQR"]
    descriptive[cname]["Greater"] = descriptive[cname]["Q3:75%"]+descriptive[cname]["1.5 Rule"]
    descriptive[cname]["Lesser"] = descriptive[cname]["Q1:25%"]-descriptive[cname]["1.5 Rule"]
    
    
    
```

```python
descriptive
```

```python
descriptive
```

```python
lesser =[]
greater=[]
    
```

```python
for cname in quan :
    if (descriptive[cname]["Min"] < descriptive[cname]["Lesser"]):
        lesser.append(cname)
    if(descriptive[cname]["Max"] > descriptive[cname]["Greater"]):
        greater.append(cname)
```

```python
for cname in lesser:
    dataset[cname][dataset[cname]<descriptive[cname]["Lesser"]] = descriptive[cname]["Lesser"]
for cname in greater:
    dataset[cname][dataset[cname]>descriptive[cname]["Greater"]] = descriptive[cname]["Greater"]
```
